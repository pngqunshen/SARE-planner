```python
def bfs(map_, start):
    neigh_func = lambda x,y: [
        (x-1,y-1), (x-1,y  ),(x-1,y+1),
        (x  ,y-1),           (x  ,y+1),
        (x+1,y-1), (x+1,y  ),(x+1,y+1)
        ]
    visited = set()
    queue = [(start[0], start[1])]
    curr = queue.pop(0)
    while map_[curr[0], curr[1]] != -1:
        # print(curr)
        neigh_lst = neigh_func(curr[0], curr[1])
        for a,b in neigh_lst:
            if not utils.out_of_bounds(a, b, map_.shape[0], map_.shape[1]):
                # print((a,b))
                # print(visited)
                if map_[a,b] != 0 and (a,b) not in visited:
                    # print('adding', (a,b), ' to queue')
                    queue.append((a,b))
                elif map_[a,b] == -1:
                    return (a,b)
            visited.add((a,b))
        curr = queue.pop(0)
```

```python
from collections import deque

def bfs_new(map_, start):
    neigh_func = lambda x,y: [
        (x-1,y-1), (x-1,y  ),(x-1,y+1),
        (x  ,y-1),           (x  ,y+1),
        (x+1,y-1), (x+1,y  ),(x+1,y+1)
        ]
    visited = set()
    queue = deque([(start[0], start[1])])
    while len(queue) != 0:
        curr = queue.popleft()
        neigh_lst = neigh_func(curr[0], curr[1])
        for a,b in neigh_lst:
            if not utils.out_of_bounds(a, b, map_.shape[0], map_.shape[1]):
                if map_[a,b] != 0 and (a,b) not in visited:
                    queue.append((a,b))
                    visited.add((a,b))
                elif map_[a,b] == -1:
                    return (a,b)
    return None
```
