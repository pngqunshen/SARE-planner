```python
from AREgym import AREEnv
import random
import os
```

```python
env = AREEnv(500, 40, save_map=True)
env.reset()
print("world generation done")
for i in range(1):
    env.step(random.random())
```

```python
img_path = "images"

if not os.path.exists(img_path):
    os.makedirs(img_path)

env.render(img_path)
```
